```python
import pandas as pd

df = pd.read_csv("data/india_housing_prices.csv")

df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates(inplace=True)

print(df.shape)
```

```python
df = df.ffill()

print(df.isnull().sum())
```

```python
df['Age_of_Property'] = 2026 - df['Year_Built']

df['Price_per_SqFt'] = (
    df['Price_in_Lakhs'] / df['Size_in_SqFt']
)

df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.histplot(df['Price_in_Lakhs'], kde=True)

plt.title("Property Price Distribution")

plt.show()
```

```python
sns.histplot(df['Size_in_SqFt'], kde=True)

plt.title("Property Size Distribution")

plt.show()
```

```python
sns.boxplot(
    x='Property_Type',
    y='Price_per_SqFt',
    data=df
)

plt.xticks(rotation=45)
plt.title("Price per SqFt by Property Type")
plt.show()
```

```python
sns.scatterplot(
    x='Size_in_SqFt',
    y='Price_in_Lakhs',
    data=df
)

plt.title("Price vs Size")

plt.show()
```

```python
sns.boxplot(df['Price_per_SqFt'])

plt.title("Outliers in Price per SqFt")
plt.show()
```

```python
state_price = df.groupby('State')['Price_per_SqFt'].mean()

state_price.sort_values(ascending=False).plot(kind='bar')

plt.title("Average Price per SqFt by State")
plt.show()
```

```python
city_price = df.groupby('City')['Price_in_Lakhs'].mean()

city_price.sort_values(ascending=False).head(10).plot(kind='bar')

plt.title("Top 10 Cities by Average Price")

plt.show()
```

```python
locality_age = df.groupby('Locality')['Age_of_Property'].median()

locality_age.sort_values(ascending=False).head(10).plot(kind='bar')

plt.title("Median Property Age by Locality")
plt.show()
```

```python
sns.countplot(
    x='BHK',
    hue='City',
    data=df
)

plt.title("BHK Distribution Across Cities")
plt.show()
```

```python
top_localities = df.groupby('Locality')['Price_in_Lakhs'] \
.mean().sort_values(ascending=False).head(5)

top_localities.plot(kind='bar')

plt.title("Top Expensive Localities")
plt.show()
```

```python
numeric_df = df.select_dtypes(include='number')

plt.figure(figsize=(12,8))

sns.heatmap(
    numeric_df.corr(),
    annot=True,
    cmap='coolwarm'
)

plt.title("Correlation Heatmap")

plt.show()
```

```python
sns.scatterplot(
    x='Nearby_Schools',
    y='Price_per_SqFt',
    data=df
)

plt.title("Schools vs Price per SqFt")
plt.show()
```

```python
sns.scatterplot(
    x='Nearby_Hospitals',
    y='Price_per_SqFt',
    data=df
)

plt.title("Hospitals vs Price per SqFt")
plt.show()
```

```python
sns.boxplot(
    x='Furnished_Status',
    y='Price_in_Lakhs',
    data=df
)

plt.title("Furnished Status vs Price")

plt.show()
```

```python
sns.boxplot(
    x='Facing',
    y='Price_per_SqFt',
    data=df
)

plt.xticks(rotation=45)
plt.title("Facing Direction vs Price per SqFt")
plt.show()
```

```python
sns.countplot(
    x='Owner_Type',
    data=df
)

plt.title("Owner Type Distribution")
plt.show()
```

```python
sns.countplot(
    x='Availability_Status',
    data=df
)

plt.title("Availability Status")
plt.show()
```

```python
sns.boxplot(
    x='Parking_Space',
    y='Price_in_Lakhs',
    data=df
)

plt.title("Parking Space vs Price")
plt.show()
```

```python
sns.boxplot(
    x='Amenities',
    y='Price_per_SqFt',
    data=df
)

plt.xticks(rotation=45)
plt.title("Amenities vs Price per SqFt")
plt.show()
```

```python
sns.boxplot(
    x='Public_Transport_Accessibility',
    y='Price_per_SqFt',
    data=df
)

plt.title("Transport Accessibility vs Price")
plt.show()
```
